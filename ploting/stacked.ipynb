## Constants 

```python
X_LABEL_FONT_SIZE = 19
Y_LABEL_FONT_SIZE = 20

X_TICKS_FONT_SIZE = 19
Y_TICKS_FONT_SIZE = 19

LEGEND_FONT_SIZE = 19

FIG_WIDTH = 10
FIG_HEIGHT = 3.5

BAR_WIDTH = 0.66

GRID_LINE_WIDTH = 0.2
GRID_LINE_COLOR = 'black'
```

## Occupancy Victima

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


X_TITLE = ''
Y_TITLE = 'Breakdown of\nL2 Cache Occupancy (%)'
Y_MIN = 0
Y_MAX = 100
Y_TICKS = 20
FILE_NAME = 'occupancy-victima'

# Read the CSV file
df = pd.read_csv(f'csvs/{FILE_NAME}.csv')

# Set index to workload
df.set_index("workload", inplace=True)

# Sort workloads alphabetically, keeping "MEAN" at the end
df = df.sort_index()
if 'MEAN' in df.index:
    mean_row = df.loc[['MEAN']]
    df = df.drop(index='MEAN')
    df = pd.concat([df, mean_row])

# Define colors (matching previous bar chart style)
colors = ["#ffffff", "#aaaaaa", "#4f4f4f"]

# Create figure and axis
fig, ax = plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT)), plt.gca()

# Plot stacked bars
df.plot(kind="bar", stacked=True, ax=ax, color=colors, edgecolor="black", linewidth=0.9, width=BAR_WIDTH, zorder=3)

# # Add hatches to first stacked bar of all workloads
# for i in range(len(df.index)):
#     for j in range(len(df.columns)):
#         if j == 0:
#             ax.patches[j * len(df.columns) + i].set_hatch('/')
#             ax.patches[j * len(df.columns) + i].set_edgecolor('black')
#             ax.patches[j * len(df.columns) + i].set_linewidth(0.9)
#         else:
#             ax.patches[j * len(df.columns) + i].set_hatch('')



# Set labels and ticks
ax.set_xlabel(X_TITLE, fontsize=X_LABEL_FONT_SIZE)
ax.set_ylabel(Y_TITLE, fontsize=Y_LABEL_FONT_SIZE)
ax.set_xticks(np.arange(len(df.index)))
ax.set_xticklabels(df.index, fontsize=X_TICKS_FONT_SIZE, rotation=45)
ax.set_yticks(np.arange(Y_MIN, Y_MAX + 0.05, Y_TICKS))
ax.set_yticklabels(np.arange(Y_MIN, Y_MAX + 0.05, Y_TICKS).astype(int), fontsize=Y_TICKS_FONT_SIZE)
ax.set_ylim(Y_MIN, Y_MAX)

# Set margins
ax.margins(x=0.02, y=0.05)

# Position the legend at the top
ax.legend(loc='upper center', frameon=False, bbox_to_anchor=(0.5, 1.24), ncol=len(df.columns), fontsize=LEGEND_FONT_SIZE)

# Add gridlines
ax.grid(axis='y', linewidth=GRID_LINE_WIDTH, color=GRID_LINE_COLOR, zorder=0)

# Save the figure
plt.savefig(f'figs/{FILE_NAME}.png', dpi=300, bbox_inches='tight', pad_inches=0.2)

# Show the plot
# plt.show()
```
