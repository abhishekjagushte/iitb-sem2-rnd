## DTLB & STLB MPKI

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

X_TITLE = ''
Y_TITLE = 'MPKI'
Y_MIN = 0
Y_MAX = 100
FILE_NAME = 'mpkis'
Y_TICKS = 20


# Read the CSV data into a DataFrame
df = pd.read_csv('csvs/' + FILE_NAME + '.csv')

# Set the workload column as the index
df.set_index('workload', inplace=True)


# Colors for the bars
colors = ["#ffffff", "#555555"]


# Create a figure and axes
fig, ax = plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT)), plt.gca()

# Plot each column as a separate bar group
bar_width = BAR_WIDTH / len(df.columns)  # Adjust bar width for spacing
x = np.arange(len(df.index))  # X locations for the groups

for i, col in enumerate(df.columns):
    ax.bar(x + i * bar_width, df[col], width=bar_width, edgecolor='black', linewidth=0.9, color=colors[i], label=col, zorder=3)

# Set the labels and ticks
ax.set_xlabel(X_TITLE, fontsize=X_LABEL_FONT_SIZE)
ax.set_ylabel(Y_TITLE, fontsize=Y_LABEL_FONT_SIZE)
ax.set_xticks(x + bar_width * (len(df.columns) - 1) / 2)
ax.set_xticklabels(df.index, fontsize=X_TICKS_FONT_SIZE, rotation=45)
ax.set_yticks(np.arange(Y_MIN, Y_MAX + 0.05, Y_TICKS))
ax.set_yticklabels(np.arange(Y_MIN, Y_MAX + 0.05, Y_TICKS).astype(int), fontsize=Y_TICKS_FONT_SIZE)
ax.set_ylim(Y_MIN, Y_MAX)


# set the margins
ax.margins(x=0.02, y=0.05)

# Position the legend at the top inside the chart area without overlapping
ax.legend(loc='upper center', frameon=False, bbox_to_anchor=(0.5, 1.2), ncol=len(df.columns), fontsize=LEGEND_FONT_SIZE)

# Add gridlines
ax.grid(axis='y', linewidth=GRID_LINE_WIDTH, color=GRID_LINE_COLOR, zorder=0)

# Save the plot with minimal white space
plt.savefig('figs/' + FILE_NAME + '.png', dpi=300, bbox_inches='tight', pad_inches=0.2)


# Show the plot
# plt.show()
```
